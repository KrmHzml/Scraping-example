```python
pages=[]
prices=[]
ratings=[]
title=[]
urls=[]


no_of_pages=1

for i in range(1,no_of_pages+1):
    url=("http://books.toscrape.com/catalogue/page-{}.html".format(i))
    pages.append(url)
```

```python
pages
```

```python
import requests
import bs4
for item in pages:
    page=requests.get(item)
    soup=bs4.BeautifulSoup(page.text,"html.parser")
    
print(soup)
```

```python
print(soup.prettify())
```

```python
for t in soup.findAll("h3"):
    print(t)
```

```python
for t in soup.findAll("h3"):
    titles=t.getText()
    title.append(titles)
```

```python
title
```

```python
for  p in soup.find_all("p",class_="price_color"):
    price=p.getText()
    prices.append(price)
```

```python
prices
```

```python
for s in soup.find_all("p",class_="star-rating"):
    for k,v in s.attrs.items():
        star=v[1]
        ratings.append(star)
        
```

```python
ratings
```

```python
divs=soup.find_all('div',class_='image_container') 

for thumbs in divs:
    tagss=thumbs.find('img',class_='thumbnail')
    links='http://books.toscrape.com/' + str(tagss['src'])
    newlinks=links.replace('..','') 
    urls.append(newlinks)
```

```python
urls
```

```python
web_data={"Title":title,
          "Price":prices,
          "Ratings":ratings,
          "URL":urls
}
```
